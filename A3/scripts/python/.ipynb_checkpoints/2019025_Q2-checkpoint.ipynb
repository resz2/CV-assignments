```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from skimage.color import label2rgb
import pandas as pd
```

```python
def euclidean(a,b):
    a = np.array(a)
    b = np.array(b)
    return np.sqrt(np.sum((a-b)**2))
```

```python
def calcMean(training, unique_labels, s):
    mean_values = []
    for i in range(s,len(unique_labels)):
        cluster = training[np.where(labels == i-s)]
        mean_values.append(np.mean(cluster, axis = 0))
    return mean_values
```

```python
img = cv.imread("dataset/0002.jpg") 
disimg = img.copy()
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
rows, cols, val = img.shape
plt.show()
```

```python
indices = np.dstack(np.indices(img.shape[:2]))
training = np.concatenate((img, indices), axis=-1) 
training = np.reshape(training, [-1,5])
```

```python
training.shape
```

```python
db = DBSCAN(eps = 11, min_samples = 50, metric = 'euclidean',algorithm ='ball_tree')
db.fit(training)
labels = db.labels_
```

```python
unique_labels = np.unique(labels)
print(unique_labels.shape)
mean_values = calcMean(training, unique_labels, 1)
```

```python
for i,label in enumerate(labels):
    if label == -1:
        dis = np.array([euclidean(mean,training[i]) for mean in mean_values])
        labels[i] = np.argmin(dis)
```

```python
unique_labels = np.unique(labels)
print(unique_labels.shape)
mean_values = calcMean(training, unique_labels, 0)
```

```python
label_image = np.reshape(labels, [rows, cols])

for i in range(label_image.shape[0]):
    for j in range(label_image.shape[1]):
#         print(i,j)
        disimg[i][j] = np.int8(mean_values[label_image[i][j]][0:3])
```

```python
plt.imshow(cv.cvtColor(disimg, cv.COLOR_BGR2RGB))
```
